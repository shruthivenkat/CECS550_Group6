```python
from sklearn.model_selection import KFold, train_test_split
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('final_features.csv')
```

```python
df.head()
```

```python
df = df.fillna('0')
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.drop(['label','kind'],axis=1)
```

```python
X.head()
```

```python
y = df['label']
```

```python
y.head()
```

```python
nb = GaussianNB()
selector = SelectKBest(score_func=mutual_info_classif, k=50)
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
selector.fit(X, y)
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
```

```python
X_selected = selector.transform(X)

# Fit the Naive Bayes model to the selected X data
nb.fit(X_selected, y)
```

```python
pipeline = Pipeline([
    ('selector', SelectKBest(score_func=mutual_info_classif, k=30)),
    ('nb', GaussianNB())
])

# Fit the pipeline on the data
pipeline.fit(X, y)

# Get the scores of the features
scores = pipeline.named_steps['selector'].scores_

# Print the top k features
k = 30
top_k_indices = scores.argsort()[::-1][:k]
print('Top', k, 'features:')
for i in top_k_indices:
    print(X.columns[i])
```
