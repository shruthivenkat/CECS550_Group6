```python
df['time_stamp'] = pd.to_datetime(df['time_stamp'], format='%m%d')
```

```python
user_activity = df.groupby('user_id')['action_type'].value_counts().unstack(fill_value=0)
```

```python
user_activity
```

```python
user_activity.columns = ['clicks', 'purchases', 'add_to_fav']
#there is no one with action type = 1 i.e. ADD TO CART, so removed
user_activity = user_activity.reset_index()
```

```python
user_activity['add_to_cart']=0
```

```python
user_activity
```

```python
merchant_activity = df.groupby('merchant_id')['action_type'].value_counts().unstack(fill_value=0)
```
